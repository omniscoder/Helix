# Edit DAG Visualization Demo

This notebook loads the bundled CRISPR edit DAG artifact, renders the graph, and shows the sequence diff between the root view and a terminal node. Everything happens entirely in silico.

```python
from pathlib import Path
import json

import matplotlib.pyplot as plt

from helix.edit.dag import dag_from_payload
from helix.visualization.dag_viz import plot_edit_dag
from helix.visualization.diff_view import genome_view_diff

artifact_path = Path("examples/crispr_edit_dag.json")
payload = json.loads(artifact_path.read_text())
dag = dag_from_payload(payload)

_ = plot_edit_dag(dag, figsize=(6, 5))
plt.show()
```

```python
root_view = dag.nodes[dag.root_id].genome_view
leaf = dag.terminal_nodes()[0]

print(genome_view_diff(root_view, leaf.genome_view, "chrDemo", context=60))
```
